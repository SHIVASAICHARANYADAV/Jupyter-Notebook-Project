```python
import sqlite3

# Connect to a new database (this will create restaurants.db)
conn = sqlite3.connect('restaurants.db')
cursor = conn.cursor()

# Open the .sql file and execute commands
with open('restaurants.sql', 'r') as f:
    sql_script = f.read()

cursor.executescript(sql_script)  # Execute all commands
conn.commit()
conn.close()

print("Database created successfully!")
```

```python
import pandas as pd
import sqlite3

# Connect to the database
conn = sqlite3.connect('restaurants.db')

# Load the restaurants table
restaurants_df = pd.read_sql_query("SELECT * FROM restaurants", conn)

# Check first 5 rows
print(restaurants_df.head())

# Close connection
conn.close()
```

```python
# Users JSON
users_df = pd.read_json('users.json')

# Orders CSV
orders_df = pd.read_csv('orders.csv')
```

```python
orders_users_df = pd.merge(orders_df, users_df, on='user_id', how='left')
```

```python
final_df = pd.merge(orders_users_df, restaurants_df, on='restaurant_id', how='left')

# Check final dataset
final_df.head()
```

```python
# Check first 10 rows
final_df.head(10)

# Check dataset size
print("Rows and Columns:", final_df.shape)

# Check for missing values
print(final_df.isnull().sum())
```

```python
# Check all column names in final dataset
print(final_df.columns)
```

```python
# Total spent by each user
total_by_user = final_df.groupby('name')['total_amount'].sum()
print("Total amount spent by each user:\n", total_by_user)
```
